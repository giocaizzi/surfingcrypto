# Reporting: figures

This module is the core for the production of the images that can be used for visually assessing the price data.

All package figures are based on `BaseFigure` object that offers basic methods for working with matplotlib figures.

```python
from surfingcrypto import Config,TS
from surfingcrypto.reporting.figures import SimplePlot,TaPlot,ATHPlot
```

Then load the configuration files.

```python
parent="/Users/giorgiocaizzi/Documents/GitHub/surfingcrypto/"
c=Config(parent+"config")
```

## Price data plots

### The `SimplePlot` figure

This is the most basic price plot.

```python
ts=TS(c,coin="BTC")
sp=SimplePlot(ts,"6m")
```

### The `TaPlot` figure

Produce a comples `ta` plot for each coin specified in the configuration file.

```python
tap=TaPlot(
    trendlines=False,
    object=ts,
    graphstart="6m"
    )
```
